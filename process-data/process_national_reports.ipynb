Volume over time by case type

```python
# Set the style
sns.set(style="whitegrid")

# Convert 'Election Held Date' to datetime
df_overall['Election Held Date'] = pd.to_datetime(df_overall['Election Held Date'])

# Define a dictionary to map Case Types to colors
dict_col = {
    'RC': 'blue',
    'RM': 'green',
    'RD': 'red',
    'UD': 'orange'
}

# Prepare the plot
plt.figure(figsize=(10, 6))

# Loop through each Case Type and plot its corresponding smoothed line
for case_type in df_overall['Case Type'].unique():
    df_filtered = df_overall[df_overall['Case Type'] == case_type]
    
    # Group by date and count the number of cases
    df_grouped = df_filtered.groupby('Election Held Date').size().reset_index(name='Count')
    
    # Sort by date (important for rolling calculations)
    df_grouped = df_grouped.sort_values('Election Held Date')
    
    # Apply rolling mean for smoothing
    df_grouped['Smoothed'] = df_grouped['Count'].rolling(window=7).mean()
    
    # Plot the smoothed data for this Case Type
    plt.plot(df_grouped['Election Held Date'], df_grouped['Smoothed'], label=f'{case_type} (7-day rolling avg.)', color=dict_col[case_type])

# Set the plot title and labels
plt.title('Volume of Cases Over Time by Case Type')
plt.xlabel('')
plt.ylabel('Number of Cases')

# Limit the x-axis to dates after 2021-01-01
plt.xlim(pd.Timestamp('2021-01-01'), pd.Timestamp('2024-06-01'))

# Add a legend and grid
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
