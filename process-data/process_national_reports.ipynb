```python
list_dfs = []
for file in os.listdir("../data/national_report_data/"):
    if file.endswith(".xlsx") and "$" not in file and "2024" in file:
        if "october" in file:
            # one org
            df = pd.read_excel("../data/national_report_data/" + file, skiprows=27, nrows=157, usecols="A:Q", 
                                names=["Region","Case Number","Case ID","Case Name","Case Type","Dispute Unit City","Dispute Unit State","Election Held Date","Num Eligible Voters","Valid Votes Against","Votes for Labor Org 1","Labor Org 1 Name","Stipulated / Consent / Directed","Cert of Rep (Win)","Cert of Results (Loss)","Closed Date","Closed Reason"])
            list_dfs.append(df)
            # two orgs
            df = pd.read_excel("../data/national_report_data/" + file, skiprows=189, nrows=5, usecols="A:T",
                                names=["Region","Case Number","Case ID","Case Name","Case Type","Dispute Unit City","Dispute Unit State","Election Held Date","Num Eligible Voters","Valid Votes Against","Votes for Labor Org 1","Labor Org 1 Name","Votes for Labor Org 2","Labor Org 2 Name","Stipulated / Consent / Directed","Cert of Rep (Win)","Cert of Results (Loss)","Union To Certify","Closed Date","Closed Reason"])
            list_dfs.append(df)
        elif "november" in file:
            # one org
            df = pd.read_excel("../data/national_report_data/" + file, skiprows=27, nrows=146, usecols="A:Q", 
                                names=["Region","Case Number","Case ID","Case Name","Case Type","Dispute Unit City","Dispute Unit State","Election Held Date","Num Eligible Voters","Valid Votes Against","Votes for Labor Org 1","Labor Org 1 Name","Stipulated / Consent / Directed","Cert of Rep (Win)","Cert of Results (Loss)","Closed Date","Closed Reason"])
            list_dfs.append(df)
            # two orgs
            df = pd.read_excel("../data/national_report_data/" + file, skiprows=181, nrows=1, usecols="A:V",
                                names=["Region","Case Number","Case ID","Case Name","Case Type","Dispute Unit City","Dispute Unit State","Election Held Date","Num Eligible Voters","Valid Votes Against","Votes for Labor Org 1","Labor Org 1 Name","Votes for Labor Org 2","Labor Org 2 Name","Votes for Labor Org 3","Labor Org 3 Name","Stipulated / Consent / Directed","Cert of Rep (Win)","Cert of Results (Loss)","Union To Certify","Closed Date","Closed Reason"])                   
            list_dfs.append(df)
            # ud elections
            df = pd.read_excel("../data/national_report_data/" + file, skiprows=187, nrows=3, usecols="A:O",
                                names=["Region","Case Number","Case ID","Case Name","Case Type","Dispute Unit City","Dispute Unit State","Election Held Date","Num Eligible Voters","For Proposition","Against Proposition","Labor Org 1 Name","Stipulated / Consent / Directed","Closed Date","Closed Reason"])
            list_dfs.append(df)      
        elif "december" in file:
            # one org
            df = pd.read_excel("../data/national_report_data/" + file, skiprows=27, nrows=145, usecols="A:Q", 
                                names=["Region","Case Number","Case ID","Case Name","Case Type","Dispute Unit City","Dispute Unit State","Election Held Date","Num Eligible Voters","Valid Votes Against","Votes for Labor Org 1","Labor Org 1 Name","Stipulated / Consent / Directed","Cert of Rep (Win)","Cert of Results (Loss)","Closed Date","Closed Reason"])
            list_dfs.append(df)
            # two orgs
            df = pd.read_excel("../data/national_report_data/" + file, skiprows=177, nrows=4, usecols="A:T",
                                names=["Region","Case Number","Case ID","Case Name","Case Type","Dispute Unit City","Dispute Unit State","Election Held Date","Num Eligible Voters","Valid Votes Against","Votes for Labor Org 1","Labor Org 1 Name","Votes for Labor Org 2","Labor Org 2 Name","Stipulated / Consent / Directed","Cert of Rep (Win)","Cert of Results (Loss)","Union To Certify","Closed Date","Closed Reason"])        
            list_dfs.append(df)
            # ud elections
            df = pd.read_excel("../data/national_report_data/" + file, skiprows=187, nrows=1, usecols="A:O",
                                names=["Region","Case Number","Case ID","Case Name","Case Type","Dispute Unit City","Dispute Unit State","Election Held Date","Num Eligible Voters","For Proposition","Against Proposition","Labor Org 1 Name","Stipulated / Consent / Directed","Closed Date","Closed Reason"])
            list_dfs.append(df)   

            
df_2024_part2 = pd.concat(list_dfs)
```
